```python
import numpy as np
import pandas as pd
import seaborn as sns
import ast, json
import missingno
from scipy import stats

from datetime import datetime
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
cars_df = pd.read_table("automobile.txt", sep=',')
cars_df = pd.DataFrame(cars_df)
print(cars_df.columns)
```

```python
cars_df.head(10)
```

```python
missingno.matrix(cars_df, figsize = (30,10))
```

```python
print(cars_df.dtypes)
```

```python
cars_df.replace('?',0,inplace=True)
cars_df['bore'] = cars_df['bore'].astype(np.float64)
cars_df['stroke'] = cars_df['stroke'].astype(np.float64)
cars_df['horsepower'] = cars_df['horsepower'].astype(np.float64)
cars_df['peak-rpm'] = cars_df['peak-rpm'].astype(np.float64)
cars_df['price'] = cars_df['price'].astype(np.float64)
```

```python
print(cars_df.dtypes)
```

```python
cars_df.head(10)
```

```python
cars_df['fuel-system'].count()
```

```python
print(cars_df.isnull().sum())
```

```python
cars_df.drop ('normalized-losses', axis=1, inplace=True)
cars_df.drop ('compression-ratio', axis=1, inplace=True)
cars_df.head(20)
```

```python
 # Code here
group_by_make = cars_df.groupby(by=['make'])

make_df = round(group_by_make.mean(),0)

engine_df = make_df.sort_values(by='highway-mpg', ascending=False)

# We will assemble a DataFrame of only the relevant features to plot  
features_of_interest = pd.DataFrame({'highway-mpg': engine_df['highway-mpg'],
                                     'city-mpg': engine_df['city-mpg'],})

## Let us plot a few of the manufactures for visibility
features_of_interest = features_of_interest.iloc[:]
features_of_interest.plot(kind='barh', figsize=(20, 10))

plt.xlabel('Mile per Gallon', fontsize=12)
plt.ylabel('Car Make', fontsize=12)
```

```python
df_bin = cars_df
fig = plt.figure(figsize=(20,10))
sns.countplot(y=df_bin['make'], data=df_bin);
```

```python
body_style_count = cars_df.groupby('body-style')['body-style'].count()
# Displaying totals.
print(body_style_count )

# 'autopct' displays the percentages on the chart
plt.figure(figsize=(6,6))
plt.title('body-style')
plt.pie(body_style_count .values, labels=body_style_count.index,autopct='%1.1f%%', textprops={'fontsize':16})
plt.show()
```

```python
group_by_fuel = cars_df.groupby(by=['fuel-system'])

fuel_df = round(group_by_fuel.mean(),0)

consumption_df = fuel_df.sort_values(by='highway-mpg', ascending=False)

# We will assemble a DataFrame of only the relevant features to plot  
features_of_interest = pd.DataFrame({'highway-mpg': consumption_df['highway-mpg'],
                                     'city-mpg': consumption_df['city-mpg'],})

## Let us plot a few of the manufactures for visibility
features_of_interest = features_of_interest.iloc[:]
features_of_interest.plot(kind='barh', figsize=(8, 5))

plt.xlabel('Mile per Gallon', fontsize=12)
plt.ylabel('Fuel System', fontsize=12)
```

```python
source - https://www.kaggle.com/code/kanncaa1/dataiteam-titanic-eda/notebook

def plot_hist(variable):
    plt.figure(figsize = (9,3))
    plt.hist(cars_df[variable], bins = 50)
    plt.xlabel(variable)
    plt.ylabel("Frequency")
    plt.title("{} distribution with hist".format(variable))
    plt.show()
```

```python
numericVar = ["horsepower", "price","highway-mpg"]
for n in numericVar:
    plot_hist(n)
```
